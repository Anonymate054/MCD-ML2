```python
import fs

INTERIM_PROJECT_DIR = fs.open_fs("../../data/interim/project")
PROCESSED_PROJECT_DIR = fs.open_fs("../../data/processed/project")
INTERIM_FILE_NAME = INTERIM_PROJECT_DIR.getsyspath("hmeq_clean.csv")
PROCESSED_FILE_NAME = PROCESSED_PROJECT_DIR.getsyspath("hmeq_processed.csv")
```

```python
import pandas as pd

data = pd.read_csv(INTERIM_FILE_NAME)
data.head()
```

### Class Balancing

```python
X = data.drop(columns=['BAD'])
y = data['BAD']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
X_resampled, y_resampled = smote.fit_resample(X, y)

df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='BAD')], axis=1)

df_resampled.head()
```

```python
df_resampled.groupby('BAD')['BAD'].count()
```

```python
df_resampled.to_csv(PROCESSED_FILE_NAME, index=False)
```
